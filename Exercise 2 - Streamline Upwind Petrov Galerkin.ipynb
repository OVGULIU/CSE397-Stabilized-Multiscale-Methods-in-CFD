```julia
# Exact solution
# righhandside
function y(x::Float64)
    z = x/L
    if (z>=0) && (z<3/8)
        A = (16*(-4 - 4exp(Pe/2) + 8exp(5Pe/8) + Pe))/((-1 + exp(Pe)) * Pe^2)
        B = -32
        C = (16*(-4 + Pe))/Pe
        D = - A
        val = A*exp(Pe*z) + B*z^2 + C*z +D
    elseif (z>=3/8) && (z<1/2)
        A =-((16*(-8 + 4*exp((7*Pe)/8) - exp((3*Pe)/8)*(-4 + Pe)))/((-1 + exp(Pe))*Pe^2))
        B = 32
        C = -8 + 64/Pe
        D = (-128 + 128*exp(Pe/2) - 256*exp((5*Pe)/8) + 16*Pe - 3*Pe^2 + exp(Pe)*(256 - 48*Pe + 3*Pe^2))/(2*(-1 + exp(Pe))*Pe^2)
        val = A*exp(Pe*(z-(3/8))) + B*(z-(3/8))^2 + C*(z-(3/8)) +D
    elseif (z>=1/2) && (z<=1.0)
        A = (16*(-4 + 8*exp(Pe/8) + exp(Pe/2)*(-4 + Pe)))/((-1 + exp(Pe))*Pe^2)
        B = 0
        C = 0
        D = 1 - A *exp(Pe/2);
        val = A*exp(Pe*(z-(1/2))) + B*(z-(1/2))^2 + C*(z-(1/2)) +D
    end
    return val
end
y(x::Vector{Float64}) = Float64[y(x[k]) for k in 1:length(x)]

# righhandside
function f(x::Float64)
    z = x/L
    if (z>=0) && (z<3/8)
        func = (16U / L) * (1 - 4z)
    elseif (z>=3/8) && (z<1/2)
        func = -(16U /L) * (2 - 4z)
    elseif (z>=1/2) && (z<=1.0)
        func = 0.0
    end
    return func
end
f(x::Vector{Float64}) = Float64[f(x[k]) for k in 1:length(x)]
```

##### Definition of linear basis functions and its piecewise constant derivatives on parent element [-1.0,1.0]

```julia
# linear basis functions and derivatives on [-1,1]
linearbasisfuns(u::Float64) = ([0.5-0.5u, 0.5+0.5u], [-0.5, 0.5])
function linearbasisfuns(u::Vector{Float64})
    Nu  = zeros(Float64,2,length(u))
    dNu = zeros(Float64,2,length(u))
    for k in 1:length(u) 
        (Nu[:,k], dNu[:,k]) = linearbasisfuns(u[k])
    end
    return (Nu,dNu)
end
```

##### Galerkin Discretization of Advection Diffusion with residual based stabilization

Galerkin Advection Diffusion with residual based stabilization goverend by function ξ(α). 
    ξ(α) = 0.0 corresponds to standard Galerkin;  ξ(α) = 1.0 to classical upwind and ξ(α) = coth(α) - (1./α) to Streamline Upwind Petrov Galerkin

```julia
# finite element assembly loop for advection diffusion with linear elements
function advectiondiffusion(Ω::Vector{Float64}, U::Float64, κ::Float64, func::Function, ξ::Function, bc::Vector{Float64})
    
    # initialize
    m = length(Ω)-1                           # number of elements
    nquad   = 2                               # number of Gauss points
    u, w    = Base.gauss(Float64,nquad)       # Gauss quadrature points and weights
    Nu, dNu = linearbasisfuns(u)              # linear basis functions and derivatives

    # allocate space for system matrix and rhs
    K = spzeros(Float64,m+1,m+1)
    F = zeros(Float64,m+1)
    
    # compute τ
    τ(κₐ::Float64,U::Float64) = sign(U)  * κₐ / U^2
    
    # loop over all elements
    for k in 1:m
        
        # map from parent element on [-1,1] to element k in physical space
        h    = Ω[k+1]-Ω[k]                   # element length
        detJ = 0.5*h                          # Jacobian determinant of affine map
        x    = mean([Ω[k+1],Ω[k]]) + u * detJ  # physical coordinates
        
        # reset element bilinear functional
        a  = zeros(Float64,2,2)                # Diffusion
        b  = zeros(Float64,2,2)                # Advection 
        r  = zeros(Float64,2,2)                # residual term
        f  = zeros(Float64,2)                  # righthandside forcing
        fᵣ = zeros(Float64,2)                  # righthandside forcing residual term
        
        # artificial diffusion κₐ w.r.t standard Galerkin / central differences 
        # and calculation of τ for element k
        κₐ = 0.5 * U * h * ξ(abs(U) * h / (2κ))
        
        # loop over quadrature points
        for i in 1:nquad

            # matrix contributions
            a += (dNu[:,i] * dNu[:,i]') * (w[i] * κ / detJ)             # Diffusion
            b += (dNu[:,i] *  Nu[:,i]') * (w[i] * U)                    # Advection
            r += (dNu[:,i] * dNu[:,i]') * (w[i] * τ(κₐ,U) * U^2 / detJ) # residual term
            
            # rhs contribution
            f +=  Nu[:,i] * (func(x[i]) * w[i] * detJ)                  # righthandside forcing
            fᵣ+= dNu[:,i] * (func(x[i]) * w[i] * τ(κₐ,U) * U)           # righthandside residual term
        end
        
        # add element contributions
        K[k:k+1,k:k+1] += a - b + r
        F[k:k+1]       += f + fᵣ
    end
    
    # imposing Dirichlet boundary conditions
    K[1,:] = 0.0; K[1,1] = 1.0; K[end,:] = 0.0; K[end,end] = 1.0
    F[1] = bc[1]; F[end] = bc[2]
    
    return K \ F
end
```

##### Compute optimal value of ξ(α) according to the "Exact artificial DIffusion Method"

```julia
# optimal ξ 
ξ(α) = coth(α) - (1./α)

# plot optimal ξ
α = linspace(0.0,30.0,200)
plot(α, ξ(α))
plot(α, (1/3)*α, "--g")
plot(α, α./α, "--g")
plot(U * h / (2κ), ξ(U * h / (2κ)),"o", label="H1 optimal xi")

# plot options
xlim([0.0,30.0]); ylim([0.0,1.2])
xlabel("alpha"); ylabel("xi")
legend(loc="upper left");
```

##### Plot results for the different finite element schemes

```julia
κₐ = 0.5 * U * h * ξ(abs(U) * h / (2κ))

# plot exact solution
xx = linspace(Ω[1],Ω[end],100)
plot(xx, y(xx),linestyle="-", color="b", label="Exact Solution");

# Galerkin
ξ(α) = 0.0
ϕ = advectiondiffusion(Ω, U, κ, f, ξ, [y(Ω[1]), y(Ω[end])]); 
plot(Ω, ϕ, linestyle="-", color="g", marker="o", label="Galerkin")

# Galerkin with exact artificial diffusion
ξ(α) = 0.0
ϕ = advectiondiffusion(Ω, U, κ+κₐ, f, ξ, [y(Ω[1]), y(Ω[end])]); 
plot(Ω, ϕ, linestyle="-", color="r", marker="o", label="Galerkin with EAD")

# SUPG
ξ(α) = coth(α) - (1./α)
ϕ = advectiondiffusion(Ω, U, κ, f, ξ, [y(Ω[1]), y(Ω[end])]); 
plot(Ω, ϕ,linestyle="-", color="y", marker="o", label="SUPG")

xlabel("X")
ylabel("Y")
title("1D Advection Diffusion")
legend(loc="upper left")
savefig("FEM_U_100_f1.pdf")
```
