```julia
using PyPlot
```

```julia
# variables
N = 10                    # number of elements
Ω = [0.0, 10.0]          # domain 
U = 50.0                 # advective velocity
Κ = 10.0                 # diffucivity coefficient

# dependent variables
h = (Ω[2]-Ω[1]) / N           # elementsize
x = linspace(Ω[1],Ω[2],N+1)   # partition
Pe = U * (Ω[2]-Ω[1]) / Κ      # Peclet number

# benchmark solution
C2 = 1 / exp(Pe); C1 = -C2
y(x) = C1 + C2 * exp(Pe * x / (Ω[2]-Ω[1]))
```

```julia
# construct central difference gradient operator
function gradient(Ω::Vector{Float64}, N::Int)
    h = (Ω[2]-Ω[1]) / N
    d = Float64[1.0 / (2*h) for i in 2:N]
    return spdiagm((-d,d),(0,2))
end

# construct central difference laplacian operator
function laplacian(Ω::Vector{Float64}, N::Int)
    h = (Ω[2]-Ω[1]) / N
    d = Float64[1.0 / h^2 for i in 2:N]
    return spdiagm((d,-2d,d),(0,1,2))
end

# construct finite difference operators
function advectiondiffusion(Ω::Vector{Float64}, y::Function, U::Float64,Κ::Float64, κ::Float64,N::Int)

    # element size
    h = (Ω[2]-Ω[1]) / N
    
    # construct finite difference operators
    ∇ = gradient(Ω,N)
    Δ = laplacian(Ω,N)

    # construct advection diffusion operator
    A = U * ∇ - (Κ + κ) * Δ
    
    # initialize numerical approximation
    y_approx = zeros(Float64,N+1); y_approx[1] = y(Ω[1]); y_approx[end] = y(Ω[2])
    
    # compute righthandside
    F = zeros(Float64,N-1)
    F -= (A[:,1]*y_approx[1] + A[:,end]*y_approx[end])
    
    # compute numerical solution
    y_approx[2:end-1] = A[:,2:end-1] \ F
    
    return y_approx
end
```

```julia
# optimal ξ 
ξ(α) = coth(α) - (1./α)

# plot optimal ξ
α = linspace(0.0,20.0,200)
plot(α, ξ(α))
plot(α, (1/3)*α, "--g")
plot(α, α./α, "--g")

xlim([0.0,20.0])
ylim([0.0,1.2])
```

```julia
# plot results for difference finite difference schemes

# plot exact solution
xx = linspace(Ω[1],Ω[2],100)
plot(xx, y(xx), linestyle="-", color="b", label="Exact Solution");

# central differences
y_central = advectiondiffusion(Ω, y, U, Κ, 0.0, N); 
plot(x,y_central,linestyle="-", color="g", marker="o", label="Central Differences");

# upwind differences
y_upwind = advectiondiffusion(Ω, y, U, Κ, 0.5*U*h, N); 
plot(x,y_upwind,linestyle="-", color="r", marker="o", label="Upwind Differences");

# exact artificial diffusion method
α = U * h / (2Κ) 
κ_art = 0.5 * U*h * ξ(α)
y_exact = advectiondiffusion(Ω, y, U, Κ, κ_art, N); 
plot(x,y_exact,linestyle="-", color="y", marker="o", label="Exact artificial diffusion");

xlabel("X")
ylabel("Y")
title("1D Advection Diffusion")
legend(loc="upper left")
```
