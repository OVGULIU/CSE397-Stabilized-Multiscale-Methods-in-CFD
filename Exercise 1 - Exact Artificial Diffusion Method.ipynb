```julia
# variables
N = 10                  # number of elements - remember number of nodes is N+1
L = 10                  # length of the domain
U = 100.0               # advective velocity
Κ = 2.0                 # diffucivity coefficient

# Load plotting tools from Python's Matplotlib library
using PyPlot

# dependent variables
h = L / N;                     println("The elementsize h is ", h)
Ω = [0.0:h:L];                   println("The mesh nodes Ω are ", Ω)
Pe = U * L / Κ;                println("The Peclet number is ", Pe)
```

```julia
# Exact solution
# y(x) = .....

# righhandside
function f(Ω::Vector{Float64},x::Float64)
    L = Ω[end] - Ω[1]
    y = x/L
    if (y>=0) && (y<3/8)
        func = (16U / L) * (1 - 4y)
    elseif (y>=3/8) && (y<1/2)
        func = -(16U /L) * (2 - 4y)
    elseif (y>=1/2) && (y<=1.0)
        func = 0.0
    end
    return func
end
f(Ω::Vector{Float64},x::Vector{Float64}) = Float64[f(Ω,x[k]) for k in 1:length(x)]
```

##### Central Differnce discretization of gradient and Laplacian operator

```julia
# construct central difference gradient operator
function gradient(Ω::Vector{Float64})
    h = Ω[2]-Ω[1]
    N = length(Ω)-1
    d = Float64[1.0 / (2*h) for i in 2:N]
    return spdiagm((-d,d),(0,2))
end

# construct central difference laplacian operator
function laplacian(Ω::Vector{Float64})
    h = Ω[2]-Ω[1]
    N = length(Ω)-1
    d = Float64[1.0 / h^2 for i in 2:N]
    return spdiagm((d,-2d,d),(0,1,2))
end
```

##### Central Difference discretization of Advection Diffusion with artificial diffusion variable κₐ 

```julia
# construct finite difference operators
function advectiondiffusion(Ω::Vector{Float64}, y::Function, f::Function, U::Float64,Κ::Float64, κₐ::Float64)

    # element size
    h = Ω[2]-Ω[1]
    N = length(Ω)-1
    
    # construct finite difference operators
    ∇ = gradient(Ω)
    Δ = laplacian(Ω)

    # construct advection diffusion operator
    A = U * ∇ - (Κ + κₐ) * Δ
    
    # initialize numerical approximation
    y_approx = zeros(Float64,N+1); y_approx[1] = y(Ω[1]); y_approx[end] = y(Ω[end])
    
    # compute righthandside
    F = f(Ω[2:end-1])
    F -= (A[:,1]*y_approx[1] + A[:,end]*y_approx[end])
    
    # compute numerical solution
    y_approx[2:end-1] = A[:,2:end-1] \ F
    
    return y_approx
end
```

##### Compute optimal value of ξ(α) according to the "Exact artificial DIffusion Method"

```julia
# optimal ξ 
ξ(α) = coth(α) - (1./α)

# plot optimal ξ
α = linspace(0.0,30.0,200)
plot(α, ξ(α))
plot(α, (1/3)*α, "--g")
plot(α, α./α, "--g")
plot(U * h / (2Κ), ξ(U * h / (2Κ)),"o", label="H1 optimal xi")

# plot options
xlim([0.0,30.0]); ylim([0.0,1.2])
xlabel("alpha"); ylabel("xi")
legend(loc="upper left");
```

##### Plot results for the different finite difference schemes

```julia
# plot exact solution
xx = linspace(Ω[1],Ω[end],100)
plot(xx, y(xx), linestyle="-", color="b", label="Exact Solution");

# central differences
y_central = advectiondiffusion(Ω, y, f, U, Κ, 0.0); 
plot(Ω,y_central,linestyle="-", color="g", marker="o", label="Central Differences");

# upwind differences
y_upwind = advectiondiffusion(Ω, y, f, U, Κ, 0.5*U*h); 
plot(Ω,y_upwind,linestyle="-", color="r", marker="o", label="Upwind Differences");

# exact artificial diffusion method
α = U * h / (2Κ) 
κ_art = 0.5 * U*h * ξ(α)
y_exact = advectiondiffusion(Ω, y, f, U, Κ, κ_art); 
plot(Ω,y_exact,linestyle="-", color="y", marker="o", label="Exact artificial diffusion");

xlabel("X")
ylabel("Y")
title("1D Advection Diffusion")
legend(loc="upper left")

# save figure in pdf format
savefig("FDM_U_100_Exact.pdf")
```
